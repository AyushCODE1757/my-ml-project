#### What are outliers?
* The data points that are significantly different from other observations.they impact:-
    * Mean and Standard Deviation
     * Regression lines
     * Model accuracy, especially in distance based algorithm(KNN, k-means)
---

#### ðŸ”¢ 5 Number Summary

* Maximum

```python
lst=[14, 56, 67, 43, 56, 79, 99, 34, 23, 100, 76, 87, 34, 86, 65, 23]
import numpy as np
min,q1,median,q3,max=np.quantile(lst,[0,0.25,0.5,0.75,1.0])
iqr=q3-q1
min,q1,median,q3,max
```

```python
lower_fence=q1-(iqr*1.5)
higher_fence=q3+(iqr*1.5)
```

```python
print(lower_fence)
print(higher_fence)
```
